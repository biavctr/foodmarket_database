# Accounts and Location

```python
samples = 1000

#Create a list of 1000 random names

pessoas = [pessoa_random().nome for i in range(samples)]

first_name = [pessoa.split()[0] for pessoa in pessoas]
last_name = [pessoa.split()[-1] for pessoa in pessoas]

#Create a list of 1000 dates between 2021-01-01 and 2022-01-01

import random
from datetime import date, timedelta

start_date = date(2021, 1, 1)
end_date = date(2023, 1, 1)

time_between_dates = end_date - start_date
days_between_dates = time_between_dates.days

random_number_of_days = [start_date + timedelta(days=random.randrange(days_between_dates)) for i in range(samples)]

#Create a list of 1000 random emails from the list of names without accents

import unidecode

first_name = [unidecode.unidecode(first_name[i]) for i in range(samples)]
last_name = [unidecode.unidecode(last_name[i]) for i in range(samples)]

emails = [f"{first_name[i].lower()}.{last_name[i].lower()}@gmail.com" for i in range(samples)]

#Create a list of 1000 random phone numbers

import random

phone_numbers = [f"({random.randint(10,99)}) {random.randint(1000,9999)}-{random.randint(1000,9999)}" for i in range(samples)]

#Creat a list of 1000 unique IDs

import uuid

ids_account = [uuid.uuid4() for i in range(samples)]


#Create dataframe

import pandas as pd

accounts = pd.DataFrame({'ACCOUNT_ID' : ids_account, 'CREATED_AT' : random_number_of_days, 'FIRST_NAME': first_name, 'LAST_NAME' : last_name,
 'EMAIL': emails, 'PHONE_NUMBER': phone_numbers})
 
accounts.head()
```

```python
#Create a list of 1000 random addresses

import random
samples = 10000

numero_address = [random.randint(1,999) for i in range(samples)]
street_address = [random.choice(['Avenida', 'Rua', 'Travessa'])+' '+pessoa_random().nome for i in range(samples)]
state_address = [random.choice(['Sao Paulo','Rio de Janeiro']) for i in range(samples)]
complement = [random.choice(['Casa','Apartamento']) + ' ' + str(random.randint(1,10)) for i in range(samples)]
cidade_address = [random.choice(['Sao Paulo','Guarulhos','Osasco','Rio de Janeiro','Niteroi']) for i in range(samples)]
cep_address = [random.randint(10000000,99999999) for i in range(samples)]

#Creat a list of 1000 unique IDs

import uuid
ids_location = [uuid.uuid4() for i in range(samples)]

location = pd.DataFrame({
    'LOCATION_ID' : ids_location,
    'STREET_NAME' : street_address,
    'STREET_NUMBER' : numero_address,
    'COMPLEMENT' : complement,
    'POSTAL_CODE' : cep_address,
    'CITY' : cidade_address,
    'STATE' : state_address,
    'COUNTRY' : ['Brasil' for i in range(samples)]
 })


location.head()
```

# Sessions

```python
samples = 10000

#Create a list of 1000 ids

import uuid

ids_session = [uuid.uuid4() for i in range(samples)]

#Sample 1000 ids_name from the list of ids_account

ids_account = accounts['ACCOUNT_ID'].sample(n=samples, replace=True, random_state=10).tolist()

#Cart id
ids_cart = [uuid.uuid4() for i in range(samples)]
random_days = [random.randint(1,360000) for i in range(samples)]
random_seconds = [random.randint(1,30*60) for i in range(samples)]

df_aux_days = pd.DataFrame({'ACCOUNT_ID' : ids_account})
df_aux_days = df_aux_days.merge(accounts[['ACCOUNT_ID','CREATED_AT']], on='ACCOUNT_ID', how='left')
df_aux_days['CREATED_AT'] = pd.to_datetime(df_aux_days['CREATED_AT'])
df_aux_days['STARTED_AT'] = df_aux_days['CREATED_AT'] + pd.to_timedelta(random_days, unit='S')
df_aux_days['ENDED_AT'] = df_aux_days['STARTED_AT'] + pd.to_timedelta(random_seconds, unit='S')
df_aux_days = df_aux_days.drop(columns=['CREATED_AT'])
df_aux_days['SESSION_ID'] = ids_session
df_aux_days['CART_ID'] = ids_cart
df_aux_days['LOCATION_ID'] = ids_location
df_aux_days = df_aux_days[['SESSION_ID','LOCATION_ID','CART_ID','ACCOUNT_ID','STARTED_AT','ENDED_AT']]



```
